```python
import pandas as pd
import numpy as np
from sklearn.linear_model import SGDClassifier
import warnings
from sklearn.metrics import classification_report, precision_score, recall_score, f1_score, confusion_matrix
from sklearn.model_selection import train_test_split

data_raw = pd.read_csv('data.csv', sep=',')
data = data_raw.copy(deep=True)
#data.columns
print(data.head())
#----------------------------------
print('*'*50)
print('Valores famhist: \n', data.famhist.value_counts())
print('*'*50)
print('Valores chd: \n', data.chd.value_counts())
#---------------------------------
famhist = data.famhist.map({'Present':0, 'Absent':1})
print('Valores famhist Transformados: \n', famhist.value_counts())
#-----------------------------------


```

# Classificacao CHD

```python
sgd = SGDClassifier(random_state=42)
chd = data.chd
for column in data.columns:
        try:
            label = data[column]
            label = label[:,np.newaxis]
            try:
                train_x, test_x, train_y, test_y = train_test_split(label,chd)
                sgd.fit(train_x, train_y)
                target_train = (train_y==1)
                target_test = (test_y==1)
                y_train_pred = sgd.predict(train_x)
                y_test_pred = sgd.predict(test_x)
                print( column.upper() ,'por CHD')
                print('='*50)
                print(classification_report(target_test, y_test_pred))
                print('*'*25)
                print('Matrix_train : \n', confusion_matrix(target_train, y_train_pred))
                print('.'*25)
                print('Matrix_test : \n', confusion_matrix(target_test, y_test_pred))
                print('*'*25)
            except:
                print('*****Erro 2*******')
        except:
            print('Erro 1')
        
```
