```python
import sklearn
import pandas as pd
from sklearn import linear_model
from sklearn import datasets
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
data_raw_1 = pd.read_csv('Advertising.csv', sep =',', index_col=None)
data_raw_2 = pd.read_csv('ex1data2.txt', sep=',', index_col=None)
data_1 = data_raw_1.copy(deep=True)
data_2 = data_raw_2.copy(deep=True)
print('*'*50)
print(data_raw_1.head())
print('*'*50)
print(data_raw_2.head())
print('*'*50)
```

```python
from sklearn.model_selection import train_test_split;
reg = linear_model.LinearRegression()

train_x, test_x, train_y, test_y = train_test_split(data_1.loc[:,['TV']], data_1.loc[:,['sales']])
reg.fit(train_x, train_y)

y_pred = reg.predict(test_x) 
#------Coeficientes----------
print('Coeficiente:', reg.coef_)
print('Erro quadratico medio: %.2f' % mean_squared_error(test_y,y_pred))
print('Coeficiente de determinacao: %.2f' % r2_score(test_y, y_pred))
#-----------plot-----------
f = plt.figure()
plt.plot(test_x,y_pred, label='Linear_Model', c='k')
plt.scatter(train_x,train_y, label='Distribuicao', c='r')
plt.legend()
plt.show()
print(train_x, train_y)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

train_x, test_x, train_y, test_y = train_test_split(data_1.loc[:,['TV']], data_1.loc[:,['sales']])

scaler.fit(train_x)
x_scaled = scaler.transform(train_x)

reg = SGDRegressor(max_iter=3, alpha=0, warm_start=True)

reg.fit(x_scaled, train_y.values.ravel())
 
    
y_pred = reg.predict(x_scaled)    
#------Coeficientes----------
#print('Coeficiente:', reg.coef_)
#print('Erro quadratico medio: %.2f' % mean_squared_error(test_y,y_pred))
#print('Coeficiente de determinacao: %.2f' % r2_score(test_y, y_pred))
#-----------plot-----------
f = plt.figure()
plt.scatter(train_x,train_y, label='Linear_Model', c='k')
plt.plot(train_x,y_pred, label='Distribuicao', c='r')
plt.legend()
plt.show()
```
